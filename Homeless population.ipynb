```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 1000)

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/FidanIlham/Downloads/Directory_Of_Homeless_Population_By_Year.csv')
```

```python
data.head()
```

```python
homeless_estimates = data.groupby('Area').sum()['Homeless Estimates'].sort_values(ascending=False)
homeless_estimates = homeless_estimates[:8,]
plt.figure(figsize=(7,5))
sns.barplot(homeless_estimates.index, homeless_estimates.values, alpha=0.8)
plt.title('Homeless Estimates')
plt.ylabel('Quantity', fontsize=12)
plt.xticks(rotation = 90)
plt.xlabel('Area', fontsize=8)
plt.show()
```

```python
homeless_estimates = data.groupby('Year').sum()['Homeless Estimates'].sort_values(ascending=False)
homeless_estimates = homeless_estimates[:8,]
plt.figure(figsize=(7,5))
sns.barplot(homeless_estimates.index, homeless_estimates.values, alpha=0.8)
plt.title('Homeless Estimates')
plt.ylabel('Quantity', fontsize=12)
plt.xticks(rotation = 90)
plt.xlabel('Year', fontsize=8)
plt.show()
```
